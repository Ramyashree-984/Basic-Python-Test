```python
class Icecream:
  def __init__(self,no_of_stock,flavour):
    self.no_of_stock=no_of_stock
    self.flavour=flavour
  def serve(self,amount):
    if self.no_of_stock>=amount:
      self.no_of_stock -= amount
      print(f"We have served your{self.flavour} icecream")
  def restock(self,quantity):
    self.no_of_stock+=quantity
    print(f"{self.flavour} icrecream is restocked")
  def check_stock(self):
      print(f"Number of items left: {self.no_of_stock}")
```

```python
ob1=Icecream(20,"Butterscotch")
ob2=Icecream(30,"Chcolate")
ob3=Icecream(40,"Venilla")
```

```python
ob1.serve(10)
ob2.serve(40)
ob3.serve(30)
ob2.restock(10)
ob1.check_stock()
```
